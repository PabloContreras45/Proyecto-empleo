```python
def limpieza_educacion(educacion):
    # Verifica si el valor es una cadena
    if isinstance(educacion, str):
        # Diccionario de reemplazo para simplificar los términos de educación
        replacements = {
            "Bachelor's degree": 'Carrera',
            "Grado": 'Carrera',
            "Licenciatura": 'Carrera',
            "Master's degree": 'Master',
            "Ingeniería": 'Carrera',
            "Ciencias de la computación": 'Carrera',
            "Estadística": 'Estadística',
            "Matemáticas": 'Matemáticas'
        }

        # Reemplazar términos en base al diccionario
        for key, value in replacements.items():
            educacion = educacion.replace(key, value)
        
        # Eliminar espacios extras
        educacion = educacion.strip()

    return educacion

# Aplica la función inicial de limpieza
df['Educación'] = df['Educación'].apply(limpieza_educacion)

# Define una función adicional para unificar combinaciones específicas
def unificar_combinaciones(educacion):
    # Verifica que el valor es una cadena antes de continuar
    if isinstance(educacion, str):
        # Ordena y ajusta términos específicos
        if 'Carrera, Master' in educacion or 'Master, Carrera' in educacion:
            educacion = 'Carrera, Master'
        elif educacion == 'Carrera, Carrera':
            educacion = 'Carrera'
        elif educacion == 'Master':
            educacion = 'Carrera, Master'
        elif sorted(educacion.split(', ')) == ['Carrera', 'Estadística', 'Matemáticas']:
            educacion = 'Carrera, Estadística, Matemáticas'
        elif sorted(educacion.split(', ')) == ['Carrera', 'Carrera', 'Master']:
            educacion = 'Carrera, Master'
    
    return educacion

# Aplica la función adicional para unificar combinaciones específicas
df['Educación'] = df['Educación'].apply(unificar_combinaciones)

# Ver los resultados agrupados
print(df['Educación'].value_counts())
```

## Exploratory Data Analysis

```python
#Outliers Salarios
# Paso 1: Limpiar y convertir los valores de "sueldo_medio" a valores numéricos
df['sueldo_medio'] = pd.to_numeric(df['sueldo_medio'].replace('[^0-9]', '', regex=True), errors='coerce')

# Paso 2: Calcular los límites de los outliers usando Tukey's Fence
q1 = df['sueldo_medio'].quantile(0.25)
q3 = df['sueldo_medio'].quantile(0.75)
iqr = q3 - q1

# Definir los límites superior e inferior
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

# Identificar los sueldos que están fuera de estos límites
df['Outlier'] = (df['sueldo_medio'] < limite_inferior) | (df['sueldo_medio'] > limite_superior)

# Paso 3: Visualizar los datos
plt.figure(figsize=(10, 6))
sns.histplot(df, x="sueldo_medio", hue="Outlier", palette={False: "blue", True: "red"}, bins=30, kde=True)
plt.title('Distribución de Sueldos Medios con Identificación de Outliers')
plt.xlabel('Sueldo Medio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
def preparacion_mapa(x):
    # Eliminar columnas innecesarias
    x.drop(['Título', 'Empresa', 'Modalidad', 'Sector', 
            'Descripción', 'Servicios', 'Otro Idioma', 'EntornoTEC', 
            'Beneficios', 'Habilidades', 'Herramientas', 'Educación'], 
        axis=1, inplace=True)
    
    # Renombrar columna 'Ubicación' a 'CCAA'
    x.rename(columns={'Ubicación': 'CCAA'}, inplace=True)
    x['CCAA'] = x['CCAA'].str.strip().str.title()  # Eliminar espacios y uniformar formato
    
    # Agregar columna 'País'
    # Diccionario de ejemplo: Comunidad Autónoma -> País
    ccaa_a_pais = {
        'Madrid': 'España',
        'Cataluña': 'España',
        'Andalucía': 'España',
        # Agregar más mapeos según sea necesario
    }
    x['País'] = x['CCAA'].map(ccaa_a_pais)

    # Agregar columna 'Continente'
    pais_a_continente = {
        'España': 'Europa',
        # Agregar más mapeos si trabajas con más países
    }
    x['Continente'] = x['País'].map(pais_a_continente)

    return x
    
```
