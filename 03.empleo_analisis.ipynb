```python
def limpieza_educacion(educacion):
    # Verifica si el valor es una cadena
    if isinstance(educacion, str):
        # Definir un diccionario de reemplazo
        replacements = {
            "Bachelor's degree": 'Carrera',
            "Grado": 'Carrera',
            "Licenciatura": 'Carrera',
            "Master's degree": 'Master',
            "Bachelor's degree, Master's degree": 'Carrera y Master',
            "Ingeniería": 'Carrera',
            "Ciencias de la computación, Ingeniería": 'Carrera',
            "Ciencias de la computación": 'Carrera',
            "Grado, Licenciatura": 'Carrera',
            "Ciencias de la computación, Estadística, Matemáticas": 'Carrera',
            "Bachelor's degree, Ciencias de la computación, Master's degree": 'Carrera y Master',
            "Ingeniería, Master's degree": 'Carrera y Master'
        }

        # Reemplazar cada clave en el diccionario por su valor
        for key, value in replacements.items():
            educacion = educacion.replace(key, value)
        
        # Aquí puedes agregar un manejo para los casos no deseados
        educacion = educacion.strip()  # Eliminar espacios al inicio y al final
        
    return educacion

# Aplica la función a la columna 'Educación'
df['Educación'] = df['Educación'].apply(limpieza_educacion)

```

```python
# Calcular el sueldo medio y agregarlo como una nueva columna
df['sueldo_medio'] = (df['sueldo_min'] + df['sueldo_max']) / 2

# Mostrar el DataFrame resultante
print(df)
```

**Exploratory Data Analysis**

```python
#Outliers Salarios
# Paso 1: Limpiar y convertir los valores de "sueldo_medio" a valores numéricos
df['sueldo_medio'] = pd.to_numeric(df['sueldo_medio'].replace('[^0-9]', '', regex=True), errors='coerce')

# Paso 2: Calcular los límites de los outliers usando Tukey's Fence
q1 = df['sueldo_medio'].quantile(0.25)
q3 = df['sueldo_medio'].quantile(0.75)
iqr = q3 - q1

# Definir los límites superior e inferior
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

# Identificar los sueldos que están fuera de estos límites
df['Outlier'] = (df['sueldo_medio'] < limite_inferior) | (df['sueldo_medio'] > limite_superior)

# Paso 3: Visualizar los datos
plt.figure(figsize=(10, 6))
sns.histplot(df, x="sueldo_medio", hue="Outlier", palette={False: "blue", True: "red"}, bins=30, kde=True)
plt.title('Distribución de Sueldos Medios con Identificación de Outliers')
plt.xlabel('Sueldo Medio')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#Relación Salario/Experiencia:

# Paso 1: Convertir la columna "sueldo_medio" a valores numéricos (ya hecho en el código anterior)
df['sueldo_medio'] = pd.to_numeric(df['sueldo_medio'].replace('[^0-9]', '', regex=True), errors='coerce')

# Paso 2: Filtrar filas con valores no nulos en "sueldo_medio" y "Expertise"
df_experiencia_salario = df.dropna(subset=['sueldo_medio', 'Expertise'])

# Paso 3: Visualizar la relación entre sueldo_medio y Expertise con un diagrama de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_experiencia_salario, x='Expertise', y='sueldo_medio', hue='Puesto', palette='viridis', s=50)
plt.title('Relación entre Sueldo Medio y Experiencia Laboral')
plt.xlabel('Años de Experiencia')
plt.ylabel('Sueldo Medio')
plt.legend(title='Puesto de Trabajo', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
