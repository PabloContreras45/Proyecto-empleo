```python
import pandas as pd
import seaborn as sns 
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datos_definitivos.csv')
```

```python
df
```

```python
df['Puesto'].value_counts()
```

```python
top_5_posiciones = ['Data Analyst', 'Data Engineer', 'Data Scientist', 'Other', 'Machine Learning Engineer']
df_top_5 = df[df['Puesto'].isin(top_5_posiciones)]
print(df_top_5['Puesto'].value_counts())
```

```python
sns.histplot(x='Puesto',data=df_top_5)
plt.xticks(rotation=60)
plt.show()
```

```python
df['Sueldo medio'].describe()
```

```python
df_top_5['Sueldo medio'].describe()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_top_5['Puesto'], y=df_top_5['Sueldo medio'], data=df_top_5, palette="Set3")
plt.title('Distribución de Sueldo Medio por Posición (Top 5)')
plt.xticks(rotation=45)
plt.show()
```
